**WHILE Loops**

```python
while condition:
    statements
    ..
    ..
```

```python
i = 1
while i < 6:
    print(i)
    i = i+1
```

```python
# BREAK statement

i = 1
while i < 6:
    print(i)
    if i == 3:
        break
    i = i + 1
```

```python
# CONTINUE statement

i = 0
while i < 6:
    i = i + 1
    if i == 3:
        continue
    print(i)
```

```python
# ELSE statement

i = 1
while i < 6:
    print(i)
    i = i + 1
else:
    print("'i' is no longer less than 6.")
```

**Lin Space** - returns even spaced numbers over specified interval

Syntax : **np.linspace(start, stop, number, endpoint, retstep, dtype, axis)

- Parameters:
- *start* : starting value of sequence
- *stop* : end value of sequence
- *number* : opt, num of samples to generate; default=50; must be non-neg
- *endpt* : opt; boolean; if True, stop is last sample; if False, stop is not included in results
- *retstep* : opt; boolean; if True, return (sample, step) where stp is spacing bt samples
- *dtype* : opt; data-type of result array; if not given, infers dtype from input args
- *axis* : optional; integer

```python
# ex is 50 numbers between 0 - 10

np.linspace(0, 10, 50)
```

**Eye** - returns array of shape, row x column; where all elements are equal to zero except the nth diagnal, whose value are equal to 1

Syntax : **np.eye(N, M=None, k=0, dtype=float, orders='C')

- Parameters:
- *N* : int; num of rows in result
- *M* : optional; integer; num of columns in result; defaults to N
- *k* : optional; integer; index (position) of the diagnal; default 0 which is main  diagnal in center
- positive integer moves diagnal index up
- negative integer moves diagnal index lower
- *dtype* : optional; data type of current array
- *order* : optional; {'C', 'F'} whether output stored as C (row) or F (col) major

```python
np.eye(4)
```

**Concatonation**

- unioning table results to the data table
- adding them together
- morph 2 sets of data together
- row does not matter but column does
- pass name of every df you want to view together
- by default, appends df in order name
- concats on row side
- similar to SQL union statement

**Merging**

Syntax : **pd.merge(var1, var2, how, on)**
- *var1* : left side
- *var2* : right side
- *how* : how to merge; default is 'inner'
- *on* : on what key; matching column name


- must define tables you want together
- each record has a primary key/unique identifier
- column joining must exist in both tables
- returns intersection of named tables
- https://www.analyticsvidhya.com/blog/2020/02/joins-in-pandas-master-the-different-types-of-joins-in-python/#:~:text=Full%20Join%20in%20Pandas,-Here's%20another%20interesting&text=Full%20Join%2C%20also%20known%20as,that%20lacks%20a%20matching%20row


- important to understand how joining bc it impacts how results look
- can lose data, get too much, or duplicate data

**Inner Merge**

- merging 2 tables together
- removes record that are not in both tables

```python
pd.merge(dfleft, dfright, how='inner', on='key')
```

```python
# passing 2 columns that match on both sides 
# pass on var a list
# default is how='inner'

pd.merge(dfleftkeys, dfrightkeys, on=['key1', 'key2'])
```

**Outer Merge**

- everything from both tables included
- returns NaNs when nothing available

```python
pd.merge(dfleftkeys, dfrightkeys, how='outer', on=['key1', 'key2'])
```

**Right Outer Merge**

- goes through right table
- then grabs everything matching with left table

```python
pd.merge(dfleftkeys, dfrightkeys, how='right', on=['key1','key2'])
```

**Left Outer Merge**

```python
pd.merge(dfleftkeys, dfrightkeys, how='left', on=['key1','key2'])
```

```python
newright = dfright.set_index('key')
newright
```

```python
# join left

newleft.join(newright)
```

```python
# outer join

newleft.join(newright, how='outer')
```
