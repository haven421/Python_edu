```python
#7. Create a function that counts the num of times a word occurs in another string.
# sentence = "Almost nothing was more annoying than having our time wasted on something not worth wasting it on."

def countWord(query, searchTerm):
    count = 0
    words = searchTerm.lower().split()
    for item in words:
        if (item == query.lower()): 
            count = count + 1 
    return count

countWord('wasted', 'Almost nothing was more annoying than having our wasted time wasted on something not worth wasting it on')
```
